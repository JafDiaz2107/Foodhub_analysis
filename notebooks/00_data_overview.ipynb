## Value Ranges and Uniqueness

```python
print(f"Unique customers: {df['customer_id'].nunique()}")
print(f"Unique restaurants: {df['restaurant_name'].nunique()}")
print(f"Unique cuisines: {df['cuisine_type'].nunique()}")
print(f"Orders: {len(df)}")
print(f"\nDay distribution:\n{df['day_of_the_week'].value_counts()}")
print(f"\nRating distribution (excluding NaN):\n{df['rating'].value_counts()}")
```

## Key Observations

- **1,898 orders** from **1,200 unique customers** across **178 restaurants** and **14 cuisines**
- **39% of orders have no rating** — significant enough to analyze as a pattern, not just missing data
- No other missing values, no duplicates
- Cost ranges from ~$4 to ~$36, prep time 20-35 min, delivery 15-33 min
