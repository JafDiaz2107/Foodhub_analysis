# 01 — Univariate EDA

Examine the distribution of each variable individually to understand central tendency, spread, and outliers.

```python
from pathlib import Path
import pandas as pd
from src.data.load import load_foodhub
from src.features.build import clean_data, engineer_features
from src.visualization.plots import (
    plot_numeric_distribution,
    plot_categorical_counts,
    plot_cuisine_distribution,
)
```

```python
df = load_foodhub(Path("data/raw/foodhub_order.csv"))
df = clean_data(df)
df = engineer_features(df)
```

## Order Cost Distribution

```python
fig = plot_numeric_distribution(df, "cost_of_the_order")
fig
```

Cost is roughly uniform between $5-$35 with no extreme outliers. The median and mean are close, suggesting relatively symmetric distribution.

## Food Preparation Time

```python
fig = plot_numeric_distribution(df, "food_preparation_time")
fig
```

Prep time centers around 27 minutes with a range of 20-35 minutes. The distribution is fairly uniform — no strong skew.

## Delivery Time

```python
fig = plot_numeric_distribution(df, "delivery_time")
fig
```

Delivery time has a wider range (15-33 min) and shows slight right skew. Some deliveries take notably longer.

## Total Time (Prep + Delivery)

```python
fig = plot_numeric_distribution(df, "total_time")
fig
```

## Cuisine Types

```python
fig = plot_cuisine_distribution(df)
fig
```

American and Japanese cuisines dominate, together accounting for over 45% of orders.

## Top Restaurants

```python
fig = plot_categorical_counts(df, "restaurant_name", top_n=15)
fig
```

Shake Shack leads with 219 orders (11.5%). The top 10 restaurants account for ~46% of all orders — a strong concentration pattern consistent with power-law distributions in marketplace platforms.

## Rating Distribution

```python
rated = df[df["rating"].notna()]
fig = plot_categorical_counts(rated, "rating")
fig
```

Among orders that received a rating, scores cluster at 4 and 5. Very few orders receive a rating of 3. This ceiling effect is common in food delivery — dissatisfied customers are more likely to skip rating entirely.
