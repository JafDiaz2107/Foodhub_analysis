# 02 — Bivariate EDA

Explore relationships between pairs of variables to find patterns that single-variable analysis can't reveal.

```python
from pathlib import Path
import pandas as pd
from src.data.load import load_foodhub
from src.features.build import clean_data, engineer_features
from src.visualization.plots import (
    plot_cost_by_cuisine,
    plot_time_by_cuisine,
    plot_weekday_weekend_comparison,
    plot_correlation_heatmap,
    plot_scatter_with_regression,
)
```

```python
df = load_foodhub(Path("data/raw/foodhub_order.csv"))
df = clean_data(df)
df = engineer_features(df)
```

## Cost by Cuisine

```python
fig = plot_cost_by_cuisine(df)
fig
```

Median cost is fairly consistent across cuisines, but the spread varies. Some cuisines (Korean, French, Italian) show wider price ranges, suggesting both budget and premium options.

## Prep and Delivery Time by Cuisine

```python
fig = plot_time_by_cuisine(df)
fig
```

Preparation time shows more cuisine-to-cuisine variation than delivery time. This makes sense — delivery depends on distance and logistics, while prep depends on the food itself.

## Weekday vs Weekend

```python
fig = plot_weekday_weekend_comparison(df)
fig
```

Average cost, prep time, and delivery time are all slightly higher on weekends. The question is whether these differences are statistically meaningful — we'll test that in notebook 03.

## Prep Time vs Cost

```python
fig = plot_scatter_with_regression(df, "food_preparation_time", "cost_of_the_order")
fig
```

No clear linear relationship between prep time and order cost. Expensive orders don't necessarily take longer to prepare.

## Prep Time vs Delivery Time

```python
fig = plot_scatter_with_regression(df, "food_preparation_time", "delivery_time")
fig
```

No strong correlation between prep and delivery times. These are largely independent processes.

## Correlation Matrix

```python
fig = plot_correlation_heatmap(df)
fig
```

The Spearman correlation matrix confirms weak correlations between most numeric variables. The strongest relationship is between `total_time` and its components (mechanical — it's a sum). No multicollinearity concerns for modeling.
