# 06 — Final Report: FoodHub Order Analysis

## Executive Summary

This analysis examined 1,898 orders from FoodHub, a food delivery platform aggregating orders from 178 restaurants across 14 cuisines. The goal was to understand demand patterns, operational bottlenecks, and customer behavior to inform business strategy.

## Key Findings

### 1. Demand Concentration
- The top 10 restaurants handle **46% of all orders** — platform revenue is heavily dependent on a small number of partners
- American and Japanese cuisines together account for **45%+ of orders**
- This concentration creates both opportunity (focus partnership efforts) and risk (partner churn impact)

### 2. Customer Retention is the Core Challenge
- **65% of customers ordered only once** and never returned
- Repeat customers spend more per order on average
- Segmentation identified 4 groups: One-Time (largest), Regular, Loyal High-Spender, and At-Risk

### 3. Rating Behavior
- **39% of orders receive no rating** — the single largest data gap
- Predictive modeling shows rating likelihood is weakly predictable from order features
- Rating absence correlates with lower engagement, not necessarily dissatisfaction

### 4. Operational Consistency
- Weekday vs weekend differences exist but are **practically small** (small effect sizes despite statistical significance)
- Food preparation time (not delivery) is the **more variable component** of total wait time
- Cuisine type drives more variation in prep time than day of week

## Methodology

| Technique | Application |
|---|---|
| Non-parametric hypothesis testing | Weekday/weekend comparisons, cuisine differences |
| Chi-square independence tests | Cuisine-day and rating-day associations |
| Rule-based segmentation (FMS) | Customer behavioral grouping |
| Binary classification (Logistic, RF, GBM) | Rating prediction with 5-fold stratified CV |
| 56 automated tests | Data validation, function correctness, model integrity |

## Recommendations

1. **Invest in retention over acquisition**: A 10% improvement in one-time customer conversion would add ~78 repeat customers — targeting them with personalized re-engagement based on their first order's cuisine preference

2. **Incentivize ratings**: The 39% gap limits analytical capability. Consider post-delivery prompts timed to when food is consumed, not when it arrives

3. **Diversify restaurant partnerships**: Top-10 concentration at 46% is a risk. Actively promote mid-tier restaurants through featured placements and promotional pricing

4. **Focus operational improvement on prep time consistency**: Delivery time variation is harder to control (depends on distance, traffic). Prep time is within restaurant control and shows the most cuisine-to-cuisine variance

## Limitations

- Dataset contains **1,898 orders** — sufficient for EDA and statistical testing, but constrains ML model complexity
- **No temporal granularity**: only weekday/weekend, no timestamps for time-of-day or seasonal analysis
- **No customer demographics**: segmentation relies solely on transactional behavior
- **Single snapshot**: no ability to track customer lifecycle or measure campaign impact over time
